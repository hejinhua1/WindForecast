```python
from dataset import CustomDataset
import plotly.express as px
import pandas as pd
import time
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from IPython.display import display, clear_output
```

```python
columns = ['10', '10_smooth', '4','5','6','7', 'predict_duration', #NWP 0-6
            'wind_record_medfilt', 'wind_record_smooth',            #wind 7-8
            'power_medfilt_unit', 'power_unit_smooth',              #power 9-10
        ]

dataset = CustomDataset("../data", "train", id=1)
# 创建 FigureWidget
fig = go.FigureWidget(make_subplots(rows=1, cols=1))

# 显示图表
display(fig)
    
# 动态刷新图表
for batch in dataset:
    df = pd.DataFrame(batch[0], columns=columns)
    df['Index'] = df.index
    
    # 清除之前的图表
    fig.data = []
    
    # 添加新的曲线图数据
    fig.add_trace(go.Scatter(x=df['Index'], y=df['10'], mode='lines', name='10'))
    fig.add_trace(go.Scatter(x=df['Index'], y=df['10_smooth'], mode='lines', name='10_smooth'))
    fig.add_trace(go.Scatter(x=df['Index'], y=df['wind_record_medfilt'], mode='lines', name='wind_record_medfilt'))
    fig.add_trace(go.Scatter(x=df['Index'], y=df['wind_record_smooth'], mode='lines', name='wind_record_smooth'))
    fig.add_trace(go.Scatter(x=df['Index'], y=df['power_medfilt_unit'], mode='lines', name='power_medfilt_unit'))
    fig.add_trace(go.Scatter(x=df['Index'], y=df['power_unit_smooth'], mode='lines', name='power_unit_smooth'))

    # 更新图表
    fig.update_layout(title='Dynamic Line Plot', xaxis_title='Index', yaxis_title='Value')
    
    # 等待 3 秒
    time.sleep(0.01)

    # 清除输出以避免重复显示
    clear_output(wait=True)
    display(fig)
```
